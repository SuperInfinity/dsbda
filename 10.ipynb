```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from PyQt6.QtWidgets.QWidget import width
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
plt.figure(figsize=(12, 8))
sns.histplot(df, x='fare', binwidth=50)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(df, x='age',y='fare')
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.displot(df, x='fare', col='sex', binwidth=50)
plt.show()
```

```python
sns.catplot(df, x='pclass', hue='survived', kind='count', col='sex')
```

```python
sns.catplot(df, x='age',hue='survived', kind='count', row='sex', height=7, aspect=3)
```
