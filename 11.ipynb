about whether they survived or not. (Column names : 'sex' and 'age')

Use :Titanic.csv

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
sns.boxplot(df, x='sex', y='age')
```

```python
sns.boxplot(df, x='sex', y='age', hue='survived')
```

```python
sns.violinplot(df, x='sex', y='age')
```

```python
sns.violinplot(df, x='sex', y='age', hue='survived' )
```

```python
sns.stripplot(df, x='sex', y='age', hue='survived')
```

```python
sns.swarmplot(df, x='sex', y='age', hue='survived')
```
